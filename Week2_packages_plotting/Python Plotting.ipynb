### A scatter plot

Now that you are familier with general plotting in any function, it's time to learn about some specific type of plots. The first one in the list is a **Scatterplot**. A scatter plot is a plot of individual data points without gives us an idea of the function it belongs too. Scatter plots have been very beneficial in fitting the data to a model, which is one of the most important aspect of the modern-day cool term *Machine Learning*.

Coming to matplotlib, there's not much difference between the syntax of a general plot and a scatter plot which will be evident from the code below: 

```python
# Defining the dataset

X_2 = np.linspace(-mt.pi,mt.pi)
Y_3 = np.sin(X_2)
```

```python
fig = plt.figure(figsize=(10, 10)) # To decide the size of the plot
fig.tight_layout()

# Checkout the line below. That's the only difference between a general plot and a scatter plot
plt.scatter(X_2,Y_3,label='y = cos(x)') 

plt.xlim(-mt.pi,mt.pi)                           # X-axis limits
plt.ylim(-1,1)                         # Y-axis limits
plt.xlabel('x')                          # X-axis label
plt.ylabel('y = g(x)')                   # Y-axis label
plt.title('A scatter plot')               # Plot title
plt.grid(visible = True)                 # Grid visibility (False by default)
plt.legend()                             # Legends
#fig = plt.figure(figsize=(6, 6))
#fig.tight_layout()
plt.show()
```

The above scatter plot is just a known function plotted in that order. This is just to show how a syntax for a scatter plot works, but that's not where it is usually used. Now, we will use a scientific dataset and see the use of a new feature

```python
date, temperature = np.loadtxt('temperatures_ghana.txt', unpack=True)

fig = plt.figure(figsize=(5, 5)) # To decide the size of the plot
fig.tight_layout()

# Checkout the line below. That's the only difference between a general plot and a scatter plot
plt.scatter(date,temperature)
plt.xlabel('Year')                          # X-axis label
plt.ylabel('Temperature')                   # Y-axis label
plt.title('A scatter plot')                 # Plot title
plt.grid(visible = True)                    # Grid visibility (False by default)
#fig = plt.figure(figsize=(6, 6))
#fig.tight_layout()
plt.show()
```

As seen above, a scatter plots looks like a bunch of points on the plot. There are some questions that can be asked for the same like is the distribution of the points uniform, or some areas have a large number of points than the other. The plot above slightly helps us in the same, but not properly because it seems a bit difficult to visualise where individual points meet.

Let's use a new paramter to see if we there's something that can be done to make the above plot better.

```python
date, temperature = np.loadtxt('temperatures_ghana.txt', unpack=True)

fig = plt.figure(figsize=(5, 5)) # To decide the size of the plot
fig.tight_layout()

# Notice the alpha parameter added for transparency
plt.scatter(date,temperature,alpha=0.6)
plt.xlabel('Year')                          # X-axis label
plt.ylabel('Temperature')                   # Y-axis label
plt.title('A scatter plot')                 # Plot title
plt.grid(visible = True)                    # Grid visibility (False by default)
#fig = plt.figure(figsize=(6, 6))
#fig.tight_layout()
plt.show()
```

This is much better for comparison and looking at the places where the heterogenity is high. 

### Something more for scientific plots: errorbars

None of the scientific data we obtain are accurate, there's always some uncertainity or error involved. It is important to represent that uncertainity too.

```python
fig = plt.figure()

x = 5*np.random.random(10) # 10 randomly spaced points between 0 and 5
y = 2.5*x + 0.8  # A simple linear equation

# Real data tends to be "noisy" - let's simulate that here

err = 0.5 # Size of errorbars - play around with this parameter!
noise = np.random.normal(scale=err, size=len(x)) # Gaussian/normal "noise"
y_data = y + noise # Adding the noise to the y-values to simulate noisy data

plt.plot(x, y, color='k')    # Makes a line, color='k' is black
plt.errorbar(x, y_data, yerr=err, fmt='.', 
             capsize=3, color='r') # Red colored points

plt.xlabel('x [arbitrary units]')
plt.ylabel('y [arbitrary units]')

# We can also set some other formatting options, like axis limits
plt.xlim(0, 5)  # x-axis range from 0 (lower limit) to 5 (upper limit)
plt.ylim(0, 12) # Same for y-axis

#plt.show()
```

## Histogram

Lets create some Gaussian-distributed (Normal) random values and plot them in a histogram. We're going to work through this code line by line and see how this affects our final plot.

If you need a refresher on what a Normal Distribution is, check out:[https://statisticsbyjim.com/basics/normal-distribution/]

```python
mean  = 50 # Mean
stdev = 10 # Standard deviation
N = 10000  # number of points to generate

x = np.random.normal(loc=mean, scale=stdev, size=N)
# random, normally distributed data points

# Histogram the data
fig, ax = plt.subplots(1,1, figsize=(8,5))
n, bins, patches = ax.hist(x, bins=50, density=1, 
                            facecolor='g', alpha=0.75)
#Play around with the hist() variables, see how they affect the plot!

ax.set(xlabel='Random value', ylabel='Probability', 
       title='Gaussian random variable', xlim=[0, 100], ylim=[0, 0.045])
       
ax.grid(True) # Shows gridlines

# Draw a vertical line at the mean and +/- 1 standard deviation
ax.axvline(mean, ls='--', color='g', lw=2)
ax.axvline(mean+stdev, ls='--', color='r', lw=2)
ax.axvline(mean-stdev, ls='--', color='r', lw=2)

fig.suptitle('A Histogram', y=1.01) # suptitle sets overall figure title
                                # The ax.set_title() is for each subplot

# Annotate the figure by placing text describing the mean and stdev
fig.text(0.6, .7, r'$\mu={:d}$, $\sigma={:d}$'.format(mean, stdev),
                    horizontalalignment='left', # abbreviate with ha=
                    verticalalignment='center')   # "  " va=
```
