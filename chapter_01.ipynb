### Functions

```python
# Example 1-6. Declaring a simple linear function
def f(x):
  return 2 * x + 1

x_vals = [0, 0.5, 1, 1.5, 2, 2.5, 3]

for x in x_vals:
  y = f(x)
  print(y)
```

```python
# Example 1-7. Plotting a linear function in Python using SymPy
from sympy import *
x = symbols('x')
f = 2 * x + 1
plot(f)
```

```python
# Example 1-8. Plotting an exponential function
x = symbols('x')
f = x**2 + 1
plot(f)
```

```python
# Example 1-9. Declaring a function with two independent variables
from sympy.plotting import plot3d

x, y = symbols('x y')
f = 2 * x + 3 * y
plot3d(f)
```
